```python
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

#Create the dataframe
df = pd.read_csv(file_url)

#Display the top 10 records
df.head()
```

```python
# Step 2: Plot the Distribution of Industry¶
# Explore how respondents are distributed across different industries.
# Plot a bar chart to visualize the distribution of respondents by industry.
# Highlight any notable trends.
```

```python
# Step 2: Plot Industry Distribution
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

# 1. Data Preparation
industry_counts = df['Industry'].value_counts().head(15)  # Top 15 industries
industry_pct = (df['Industry'].value_counts(normalize=True).head(15) * 100).round(1)

# 2. Create Visualization
ax = sns.barplot(
    y=industry_counts.index,
    x=industry_counts.values,
    palette="viridis",
    orient='h'
)

# 3. Add Value Annotations
for i, (value, pct) in enumerate(zip(industry_counts.values, industry_pct.values)):
    ax.text(value + 0.02*max(industry_counts),  # Position slightly right of bar
            i,  # Same y-position as bar
            f'{value:,} ({pct}%)',  # Format: count (percentage)
            va='center',
            fontsize=10)

# 4. Customize Plot
plt.title('Top 15 Industries by Respondent Count', pad=20, fontsize=16)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Industry', fontsize=12)
plt.xlim(0, max(industry_counts) * 1.15)  # Add padding for annotations

# 5. Add Insights Box
insights = """
Key Observations:
- Top Industry: {}
- Contains {:.1f}% of all respondents
- Top 5 industries account for {:.1f}%
""".format(
    industry_counts.index[0],
    industry_pct.iloc[0],
    industry_pct.head(5).sum()
)

plt.gcf().text(0.95, 0.5, insights,
              bbox=dict(facecolor='white', alpha=0.8),
              fontsize=11,
              ha='left')

plt.tight_layout()
plt.show()

# 6. Additional Analysis
print("\nIndustry Distribution Statistics:")
print("-" * 40)
print(f"Total industries represented: {df['Industry'].nunique()}")
print(f"Industries with <10 respondents: {(df['Industry'].value_counts() < 10).sum()}")
```

```python
# Step 3: Identify High Compensation Outliers

# Identify respondents with extremely high yearly compensation.

# Calculate basic statistics (mean, median, and standard deviation) for ConvertedCompYearly.

# Identify compensation values exceeding a defined threshold (e.g., 3 standard deviations above the mean).
```

```python
# Step 3: Identify High Compensation Outliers
import numpy as np

# 1. Clean compensation data
df_comp = df.dropna(subset=['ConvertedCompYearly']).copy()
df_comp['ConvertedCompYearly'] = pd.to_numeric(df_comp['ConvertedCompYearly'], errors='coerce')
df_comp = df_comp[df_comp['ConvertedCompYearly'] > 0]  # Remove zero/negative values

# 2. Calculate statistics
stats = {
    'Mean': df_comp['ConvertedCompYearly'].mean(),
    'Median': df_comp['ConvertedCompYearly'].median(),
    'Std Dev': df_comp['ConvertedCompYearly'].std(),
    'Min': df_comp['ConvertedCompYearly'].min(),
    'Max': df_comp['ConvertedCompYearly'].max()
}

# 3. Define outlier threshold (3 std devs above mean)
threshold = stats['Mean'] + 3 * stats['Std Dev']
outliers = df_comp[df_comp['ConvertedCompYearly'] > threshold]

# 4. Visualization
plt.figure(figsize=(14, 6))

# Boxplot with logarithmic scale
plt.subplot(1, 2, 1)
sns.boxplot(x=df_comp['ConvertedCompYearly'], showfliers=False)
plt.title('Compensation Distribution (Without Outliers)', fontsize=12)
plt.xlabel('Yearly Compensation (USD)', fontsize=10)

# Highlight outliers
plt.subplot(1, 2, 2)
sns.scatterplot(
    x=range(len(outliers)),
    y=outliers['ConvertedCompYearly'],
    color='red',
    s=100,
    alpha=0.6
)
plt.axhline(y=threshold, color='green', linestyle='--', label=f'Threshold: ${threshold:,.0f}')
plt.title('Outliers Detection', fontsize=12)
plt.ylabel('Yearly Compensation (USD)', fontsize=10)
plt.xlabel('Respondent Index', fontsize=10)
plt.legend()
plt.tight_layout()

# 5. Display results
print("Compensation Statistics:")
print("-" * 40)
for k, v in stats.items():
    print(f"{k}: ${v:,.2f}")

print(f"\nOutlier Threshold (3σ): ${threshold:,.2f}")
print(f"Number of outliers: {len(outliers):,} ({len(outliers)/len(df_comp):.2%})")

print("\nTop 5 Outliers:")
display(outliers.nlargest(5, 'ConvertedCompYearly')[['Country', 'Employment', 'YearsCodePro', 'ConvertedCompYearly']])

# 6. Advanced outlier analysis (optional)
Q1 = df_comp['ConvertedCompYearly'].quantile(0.25)
Q3 = df_comp['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
alt_threshold = Q3 + 3 * IQR

print(f"\nAlternative Threshold (3*IQR): ${alt_threshold:,.2f}")
print(f"Outliers by IQR method: {len(df_comp[df_comp['ConvertedCompYearly'] > alt_threshold]):,}")
```

```python
# Step 4: Detect Outliers in Compensation

# Identify outliers in the ConvertedCompYearly column using the IQR method.

# Calculate the Interquartile Range (IQR).

# Determine the upper and lower bounds for outliers.

# Count and visualize outliers using a box plot.
```

```python
# Step 4: Detect Compensation Outliers with IQR Method
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Preparation
comp_clean = df['ConvertedCompYearly'].dropna()  # Remove missing values
comp_clean = comp_clean[comp_clean > 0]  # Remove zero/negative compensation

# 2. Calculate IQR Statistics
Q1 = np.percentile(comp_clean, 25)
Q3 = np.percentile(comp_clean, 75)
IQR = Q3 - Q1
median = np.median(comp_clean)

# 3. Define Outlier Bounds (1.5*IQR is standard, 3*IQR for extreme outliers)
lower_bound = Q1 - 3 * IQR  # Using 3xIQR for more conservative detection
upper_bound = Q3 + 3 * IQR

# 4. Identify Outliers
outliers = comp_clean[(comp_clean < lower_bound) | (comp_clean > upper_bound)]
normal_values = comp_clean[(comp_clean >= lower_bound) & (comp_clean <= upper_bound)]

# 5. Create Enhanced Box Plot Visualization
plt.figure(figsize=(12, 8))

# Main boxplot without outliers (we'll add them separately)
boxprops = dict(linestyle='-', linewidth=2, color='darkblue')
flierprops = dict(marker='', markersize=0)  # Hide default outliers

sns.boxplot(x=normal_values,
            color='lightblue',
            width=0.4,
            boxprops=boxprops,
            flierprops=flierprops)

# Add custom outlier points
y = np.random.normal(0, 0.01, size=len(outliers))  # Jitter for visibility
plt.scatter(outliers, y,
           color='red',
           alpha=0.5,
           s=100,
           label=f'Outliers (n={len(outliers)})')

# Add reference lines and annotations
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound (${upper_bound:,.0f})')
plt.axvline(x=lower_bound, color='orange', linestyle='--', label=f'Lower Bound (${lower_bound:,.0f})')
plt.axvline(x=median, color='purple', linestyle='-', label=f'Median (${median:,.0f})')

# Formatting
plt.title('Yearly Compensation Distribution with Outlier Detection\n(IQR Method with 3×IQR Threshold)',
          fontsize=14, pad=20)
plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.yticks([])  # Hide y-axis as it's meaningless for boxplot
plt.xscale('log')  # Log scale for better visualization
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x', alpha=0.3)

# Add summary table
stats_text = f"""
Statistics Summary:
- Q1 (25th %ile): ${Q1:,.0f}
- Q3 (75th %ile): ${Q3:,.0f}
- IQR: ${IQR:,.0f}
- Median: ${median:,.0f}
- Normal Range: ${lower_bound:,.0f} to ${upper_bound:,.0f}
- Outliers: {len(outliers):,} ({len(outliers)/len(comp_clean):.1%})
"""
plt.gcf().text(0.95, 0.5, stats_text,
              bbox=dict(facecolor='white', alpha=0.8),
              fontsize=11)

plt.tight_layout()
plt.show()

# 6. Outlier Analysis
print("\nTop 5 Highest Outliers:")
print(outliers.nlargest(5).apply(lambda x: f"${x:,.0f}").to_string(index=False))

print("\nTop 5 Lowest Outliers:")
print(outliers.nsmallest(5).apply(lambda x: f"${x:,.0f}").to_string(index=False))

# 7. Optional: Create outlier dataframe with full records
outlier_records = df[df['ConvertedCompYearly'].isin(outliers)]
print(f"\nFound {len(outlier_records)} outlier records")
```

```python
# Step 5: Remove Outliers and Create a New DataFrame

# Remove outliers from the dataset.

# Create a new DataFrame excluding rows with outliers in ConvertedCompYearly.

# Validate the size of the new DataFrame.
```

```python
# Step 5: Remove Outliers and Create Clean DataFrame
import numpy as np

# 1. Prepare clean compensation data (handle missing/negative values)
df_clean = df.dropna(subset=['ConvertedCompYearly']).copy()
df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
df_clean = df_clean[df_clean['ConvertedCompYearly'] > 0]  # Remove zero/negative compensation

# 2. Calculate IQR bounds (using 3xIQR for conservative outlier removal)
Q1 = df_clean['ConvertedCompYearly'].quantile(0.25)
Q3 = df_clean['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 3 * IQR
upper_bound = Q3 + 3 * IQR

# 3. Create new DataFrame without outliers
df_no_outliers = df_clean[
    (df_clean['ConvertedCompYearly'] >= lower_bound) &
    (df_clean['ConvertedCompYearly'] <= upper_bound)
].copy()

# 4. Validation and Reporting
original_size = len(df)
clean_size = len(df_clean)
final_size = len(df_no_outliers)
outliers_removed = clean_size - final_size

print("Data Cleaning Report:")
print("-" * 40)
print(f"Original dataset size: {original_size:,} rows")
print(f"After cleaning missing/negative values: {clean_size:,} rows")
print(f"Outlier bounds: ${lower_bound:,.0f} to ${upper_bound:,.0f}")
print(f"Outliers removed: {outliers_removed:,} rows ({outliers_removed/clean_size:.1%})")
print(f"Final clean dataset size: {final_size:,} rows ({final_size/original_size:.1%} of original)")

# 5. Compare statistics
print("\nCompensation Statistics Comparison:")
print("-" * 40)
stats_comparison = pd.DataFrame({
    'Original': df_clean['ConvertedCompYearly'].describe(),
    'No Outliers': df_no_outliers['ConvertedCompYearly'].describe()
})
display(stats_comparison.style.format('{:,.2f}'))

# 6. Visual Verification
plt.figure(figsize=(12, 6))
sns.boxplot(data=[df_clean['ConvertedCompYearly'], df_no_outliers['ConvertedCompYearly']],
           palette=['lightcoral', 'lightgreen'])
plt.xticks([0, 1], ['With Outliers', 'Without Outliers'])
plt.title('Compensation Distribution Before/After Outlier Removal', fontsize=14)
plt.ylabel('Yearly Compensation (USD)')
plt.yscale('log')  # Log scale for better visualization
plt.grid(axis='y', alpha=0.3)
plt.show()

# 7. Save the new DataFrame (optional)
# df_no_outliers.to_csv('survey_data_no_outliers.csv', index=False)
```

```python
# Step 6: Correlation Analysis¶

# Analyze the correlation between Age (transformed) and other numerical columns.

# Map the Age column to approximate numeric values.

# Compute correlations between Age and other numeric variables.

# Visualize the correlation matrix.
```

```python
# Step 6: Correlation Analysis with Age
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Preparation
df_corr = df.copy()

# Transform Age to numerical values (handling categorical ranges)
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# Apply mapping only to existing categories
existing_ages = {k: v for k, v in age_mapping.items() if k in df_corr['Age'].unique()}
df_corr['Age_Numeric'] = df_corr['Age'].map(existing_ages)

# 2. Select numerical columns for correlation analysis
numerical_cols = df_corr.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Age_Numeric' not in numerical_cols:
    numerical_cols.append('Age_Numeric')

# Ensure we have valid numerical data
df_numerical = df_corr[numerical_cols].dropna()

# 3. Compute correlation matrix
corr_matrix = df_numerical.corr(method='pearson')

# 4. Visualize correlation matrix with Age focus
plt.figure(figsize=(12, 10))

# Create mask to highlight Age correlations
age_corrs = corr_matrix['Age_Numeric'].sort_values(ascending=False)
age_corrs = age_corrs.drop('Age_Numeric')  # Remove self-correlation

# Plot top correlations with Age
plt.subplot(2, 1, 1)
sns.barplot(x=age_corrs.values, y=age_corrs.index, palette='coolwarm')
plt.title('Top Correlations with Age', fontsize=14)
plt.xlabel('Pearson Correlation Coefficient')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid(axis='x', alpha=0.3)

# Plot full correlation matrix heatmap
plt.subplot(2, 1, 2)
sns.heatmap(corr_matrix,
           annot=True,
           fmt='.2f',
           cmap='coolwarm',
           center=0,
           vmin=-1,
           vmax=1,
           linewidths=0.5)
plt.title('Numerical Variables Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()

# 5. Statistical summary
print("Age Correlation Analysis Summary:")
print("-" * 40)
print(f"Strongest positive correlation: {age_corrs.idxmax()} ({age_corrs.max():.2f})")
print(f"Strongest negative correlation: {age_corrs.idxmin()} ({age_corrs.min():.2f})")
print("\nTop 5 Correlations with Age:")
display(age_corrs.head(5).to_frame('Correlation'))
print("\nBottom 5 Correlations with Age:")
display(age_corrs.tail(5).to_frame('Correlation'))

# 6. Pairplot for top correlated variables (optional)
top_3_pos = age_corrs.head(3).index.tolist()
top_3_neg = age_corrs.tail(3).index.tolist()

if len(top_3_pos) > 0 and len(top_3_neg) > 0:
    sns.pairplot(df_numerical[['Age_Numeric'] + top_3_pos + top_3_neg],
                diag_kind='kde',
                plot_kws={'alpha': 0.6})
    plt.suptitle('Pairwise Relationships with Age', y=1.02)
    plt.show()
```
